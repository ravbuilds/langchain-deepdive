### Streaming & Batch
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. 
```
stream() --> returns an iterator that yields output chunks as they are produced.
```

**invoke vs stream**<br/>
**invoke()** -> This method waits for the whole response to generate and then spits out everything.<br/>
**stream()** -> This method does not waits for the whole response to generated, instead it output what is generated so far


````python
# Normal invoke method -> This waits for the whole response to be generated and then print 
model.invoke("Write a 100 words paragraph on AI.")
````

````python
# Streaming -> This output a generator object
model.stream("write a 100 word paragraph on NLP.")
````

````python
for chunk in model.stream("write a 300 word paragraph on NLP."):
    print(chunk.text, end="|", flush=True)
````

**Batch**
Batch is a collection of independent requests to a model that can significantly improve performance and reduce costs, as the processing can be done in parallel.

````python
responses = model.batch([
    "How is NLP related to LLM?",
    "What is the LLM for Computer Vision model",
    "What is Quantum Computing?"
], config={'max_concurrency': 5})
for response in responses:
    print(response.content)
````
