## Model Integration - OpenAI, Gemini and Groq

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
# API Keys
os.environ["OPENAI_API_KEY"] = os.getenv("API_KEY_OPENAI")
os.environ["GEMINI_API_KEY"] = os.getenv("API_KEY_GEMINI")
os.environ["GROQ_API_KEY"] = os.getenv("API_KEY_GROQ")
```

### OpenAI GPT Model

```python
# Get Model
from langchain.chat_models import init_chat_model #chat_completion pattern
model = init_chat_model("gpt-4.1")
model
```

```python
# Invoke the model
response = model.invoke("Hello how are you?")
response
```

```python
#### Create model using ChatOpenAI
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model="gpt-4.1")
model
```

```python
# Run the model
response = model.invoke("Hello Howdy?")
response
```
